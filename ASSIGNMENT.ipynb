```python
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
plt.bar(data['Classification'],data['SL'])
plt.xlabel('Classification')
plt.ylabel('SL')
plt.title('plot of Classification VS SL')

plt.subplot(2,2,2)
plt.bar(data['Classification'],data['SW'])
plt.xlabel('Classification')
plt.ylabel('SW')
plt.title('plot of Classification VS SW')

plt.subplot(2,2,3)
plt.bar(data['Classification'],data['PL'])
plt.xlabel('Classification')
plt.ylabel('PL')
plt.title('plot of Classification VS PL')
plt.subplot(2,2,4)
plt.bar(data['Classification'],data['PW'])
plt.xlabel('Classification')
plt.ylabel('PW')
plt.title('plot of Classification VS PW')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,10))
plt.subplot(3,2,1)
plt.scatter(data['SL'],data['PL'])
plt.xlabel('SL')
plt.ylabel('PL')
plt.title('plot of SL VS PL')

plt.subplot(3,2,2)
plt.scatter(data['SL'],data['PW'])
plt.xlabel('SL')
plt.ylabel('PW')
plt.title('plot of SL VS PW')

plt.subplot(3,2,3)
plt.scatter(data['SW'],data['PL'])
plt.xlabel('SW')
plt.ylabel('PL')
plt.title('plot of SW VS PL')

plt.subplot(3,2,4)
plt.scatter(data['SW'],data['PW'])
plt.xlabel('SW')
plt.ylabel('PW')
plt.title('plot of SW VS PW')

plt.subplot(3,2,5)
plt.scatter(data['PL'],data['PW'])
plt.xlabel('PL')
plt.ylabel('PW')
plt.title('plot of PL VS PW')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x='SL',y='SW',hue='Classification',data=data)
plt.xticks(rotation=90)
plt.title('plot of SL VS SW')

plt.subplot(2,2,2)
sns.boxplot(x='PL',y='PW',hue='Classification',data=data)
plt.xticks(rotation=90)
plt.title('plot of PL VS PW')
plt.tight_layout()
plt.show()
```
